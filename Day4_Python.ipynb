```python
#user defined exception
class TooYoungException(Exception):
  def __init__(self,arg):
    self.msg=arg

class TooOldException(Exception):
  def __init__(self,arg):
    self.msg=arg

age=int(input("Enter you age"))
if age>60:
  raise TooYoungException('Wait sometime ,we will find best match for you')
elif age<18:
  raise TooOldException("You are not eligible for marry")
else:
  print("You are eligible for marry ")
```

```python
#palindrome
def Palindrome():
  s=[1,2,3,2,1]
  #s=[4,5,6,5,4]
  b=[]
  c=0
  for i in range(len(s)-1,-1,-1):
    b.append(s[i])
  print(b)
  for i in range(0,len(b)):
    if s[i]==b[i]:
      c=c+1
    else:
      print("One element found different")
      break
  if c==len(s):
    print("palindrome")

Palindrome()
```

```python
#merging
def merging():
  a=[1,3,5]
  b=[2,4,6]
  c=[]
  for i in range(0,3):
    for j in range(0,3):
      if a[i]<b[j]:
        if a[i] not in c:
          c.append(a[i])
      else:
        if b[j] not in c:
          c.append(b[j])
  print(c)

merging()
```

```python
#Object Oriented Programming
# Class is a logical representation of things ,cLASS IS LIKE A Blueprint or template for the object
# Object is the physical representation of class

class Student:
  def __init__(self):
    self.name='Ashish'
    self.rollno=101
    self.percentage=85

  def showInfo(self):
    print("Name",self.name)
    print("rollno",self.rollno)
    print("percentage",self.percentage)

obj=Student()
obj.showInfo()
```

```python
class Student:
  def __init__(self,name,age,percentage):
    self.name=name
    self.age=age
    self.percentage=percentage

  def showInfo(self):
    print("Name: ",self.name)
    print("Age: ",self.age)
    print("Percentage: ",self.percentage)

s=Student("Joy",22,55)
s1=Student("aditya",22,55)
s.showInfo()
s1.showInfo()
```
